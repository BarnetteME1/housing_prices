The below formula is how the competition is being scored. It's being done by taking the root mean squared error of the log. This should give me an idea of how well my regressions doing.

```python
def rmse_log(value,prediction):
    return np.sqrt(mean_squared_error(np.log(value),np.log(prediction)))
```
